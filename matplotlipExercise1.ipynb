```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# DATA
x = ["1.USA", "2.CHINA", "3.JAPAN", "4.GERMANY", "5.UK", "6.FRANCE"]
y = [20410, 14090, 5170, 4210, 2940, 2930]
z = [63039, 9957, 40649, 51158, 44162, 44915]
```

```python
# plt.plot(z, x, "r")
# plt.plot(x, y, "b")
plt.plot(x, z, "r")

# eksen balklar
plt.xticks(rotation=45)
plt.xlabel("COUNTRIES")
plt.ylabel("GDP PER CAPITA")
plt.title("GDP PER CAPITA BY COUNTRY")
```

```python
# cascading graphs:
# 1 row 2 col graph, first graph is:
plt.subplot(1, 2, 1)
plt.plot(x, z, "y*-")
plt.xticks(rotation=45)
# 1 row 2 col graph, second graph is:
plt.subplot(1, 2, 2)
plt.plot(x, y, "r--")
plt.xticks(rotation=60)
# skkl gidermek iin
plt.tight_layout()
```

```python
# OOP style
f = plt.figure()
axes = f.add_axes([0.3, 0.1, 1.2, 0.8])
axes.plot(x, y)

axes.set_xlabel("LKELER")
axes.set_ylabel("MLL GELR")
axes.set_title("LKELERE GRE MLL GELR")
```

```python
# second method
f2 = plt.figure()
axes1 = f2.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = f2.add_axes([0.56, 0.64, 0.4, 0.3])

axes1.plot(x, y, "y")
axes2.plot(x, z, "r")

# Rotate x-labels of axes2
axes2.set_xticklabels(x, rotation=30)
# plt.xticks(rotation=45)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, "y")
axes[0].set_title("MLL GELR")
axes[0].set_xticklabels(x, rotation=30)

axes[1].plot(x, z, "r")
axes[1].set_title("K BAINA")
axes[1].set_xticklabels(x, rotation=30)

plt.tight_layout()
```

```python
f3 = plt.figure()
axes = f3.add_axes([0.1, 0.1, 0.9, 0.9])

axes.plot(x, y, color="pink", linewidth=2.5, linestyle="--", marker="o", markersize=8.5, 
          markerfacecolor="aqua", markeredgewidth=3, markeredgecolor="blue", alpha=0.9)
axes.set_xlabel("LKELER")
axes.set_ylabel("MLL GELR")
axes.set_title("LKELERE GRE MLL GELR")
```

```python
# For limited outputs:

f4 = plt.figure(dpi=120)
axes = f4.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x, y, color="aqua", linewidth=3, linestyle="--")

axes.set_xlabel("LKELER")
axes.set_ylabel("MLL GELR")
axes.set_title("LKELERE GRE MLL GELR")

# limiting the axis values:

axes.set_xlim([2, 4])
axes.set_ylim([1000, 7000])
```

```python
f4 = plt.figure(figsize=(5, 2))
axes = f4.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x, y, color="purple", linewidth=3, linestyle="--")

axes.set_xlabel("LKELER")
axes.set_ylabel("MLL GELR")
axes.set_title("LKELERE GRE MLL GELR")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6,3))

axes[0].plot(x, y, "b")
axes[0].set_title("MLL GELR")

axes[1].plot(x, z, "g")
axes[1].set_title("K BAINA")

plt.tight_layout()
```
