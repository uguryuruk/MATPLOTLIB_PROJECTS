```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x = ["1.ABD", "2.ÇİN", "3.JAPONYA", "4.ALMANYA", "5.İNGİLTERE", "6.FRANSA"]
y = [20410, 14090, 5170, 4210, 2940, 2930]
z = [63039, 9957, 40649, 51158, 44162, 44915]
```

```python
# plt.plot(x,y)
plt.plot(x,z,"r")

# eksen başlıkları
plt.xlabel("ÜLKELER")
plt.ylabel("MİLLİ GELİR")
plt.title("ÜLKELERE GÖRE MİLLİ GELİR")

# Gösterim için
plt.show()
```

```python
# içi içe grafikler:
# 1 satır iki sütun grafik, ilki için:
plt.subplot(1, 2, 1)
plt.plot(x, y, "r--")
# 1 satır iki sütun grafik, ikincisi için:
plt.subplot(1, 2, 2)
plt.plot(x, z, "y*-")

# sıkışıklığı gidermek için
plt.tight_layout()
```

```python
# ikinci şekilde yapmak için
f = plt.figure()
axes = f.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x, y)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
# iç içe ikinci şekilde yapmak için
# TODO: değerlerle oyna!
f2 = plt.figure()
axes1 = f2.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = f2.add_axes([0.55, 0.55, 0.4, 0.4])

axes1.plot(x, y, "y")
axes2.plot(x, z, "r")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, "b")
axes[0].set_title("MİLLİ GELİR")

axes[1].plot(x, z, "g")
axes[1].set_title("KİŞİ BAŞINA")

plt.tight_layout()
```
